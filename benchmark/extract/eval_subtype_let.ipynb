# Evaluate performance of using let expressions in Î£-types

```python
import pandas as pd

df_valid = pd.read_csv('./data/subtype_extract_valid/ufbv.csv', sep=';').filter(['MAXLEN', 'TIME'])
valid_agg = df_valid.groupby('MAXLEN').agg(['min', 'max', 'mean', 'std'])

df_let = pd.read_csv('./data/subtype_extract_let/ufbv.csv', sep=';').filter(['MAXLEN', 'TIME'])
let_agg = df_let.groupby('MAXLEN').agg(['min', 'max', 'mean', 'std'])

ax = valid_agg.loc[:, ('TIME', 'mean')].plot(style='.', figsize = (40,10), grid=True, logy=False)
let_agg.loc[:, ('TIME', 'mean')].plot(style='.', figsize = (40,10), grid=True, logy=False, ax=ax)
```
