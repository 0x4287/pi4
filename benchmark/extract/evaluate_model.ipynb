The following plots show how long it takes Z3 to find a model for the encoding of the following subtype and supertype respectively for varying maximum bit lengths.
```
{z:⊤
  | 
    z.A.valid ∧ 
    z.B.valid ∧ 
    z.pkt_in.length + 4 > 5 ∧
    z.A@z.pkt_in = x.pkt_in ∧ 
    z.pkt_out = x.pkt_out ∧ 
    x.A.valid => z.A.valid ∧ z.A = x.A ∧
    x.B.valid => z.B.valid ∧ z.B = x.B }
<: 
Σy:A.B
```

Variable *x* is bound to type `{y:ε|y.pkt_in.length > 5}`.

````python
import pandas as pd
import matplotlib.ticker as plticker
````

## Subtype

````python
df = pd.read_csv('./data/model/ufbv/subtype.csv', sep=';').filter(['MAXLEN', 'TIME'])
agg = df.groupby('MAXLEN').agg(['min', 'max', 'mean', 'std'])
````

### Mean execution time

````python
ax = agg.loc[:, ('TIME', 'mean')].plot(style='.',figsize = (40,10), grid=True)
ax.xaxis.set_major_locator(plticker.MultipleLocator(25.0))
ax.set_xlim(0)
ax.set_xlabel('Maxlen')
ax.set_ylabel('Time (ms)')
````

### Standard Deviation


````python
ax = agg.loc[:, ('TIME', 'std')].plot(style='.', figsize = (40,10), grid=True, logy=False)
ax.xaxis.set_major_locator(plticker.MultipleLocator(25.0))
ax.set_xlim(0)
ax.set_xlabel('Maxlen')
ax.set_ylabel('Time (ms)')
````
